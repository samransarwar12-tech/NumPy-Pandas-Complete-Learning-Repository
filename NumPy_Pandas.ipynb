```python
import numpy as np
```

## Creating Arrays

```python
arr1=np.array([1,2,3])
arr2=np.array((4,5,6))
arr3=np.array([[1,2],[3,4]])
arr4=np.array([True,False,True])
print(arr1,arr2,arr3,arr4)
```

## Array Dimensions

```python
print(np.array([1,2,3]).ndim)
print(np.array([[1,2],[3,4]]).ndim)
print(np.array([[[1],[2]],[[3],[4]]]).ndim)
print(np.array([[[[1]]]]).ndim)
```

## Shape & Reshape

```python
a=np.arange(6)
print(a.shape)
print(a.reshape(2,3))
print(a.reshape(3,2))
print(a.reshape(1,6))
print(a.reshape(1,4))
```

## Data Types

```python
print(np.array([1,2,3]).dtype)
print(np.array([1.1,2.2]).dtype)
print(np.array([1,2,3],dtype=float).dtype)
print(np.array(['ABVFJDDH','BCJ']).dtype)
```

## Arithmetic Operations

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

## Comparison Operations

```python
a=np.array([10,20,30])
print(a>15)
print(a==20)
print(a!=20)
print(a<=20)
```

## Broadcasting

```python
a=np.array([1,2,3])
print(a+10)
print(a*5)
print(a+[10,20,30])
print(np.array([[1],[2],[3]])+np.array([10,20]))
```

## zeros ones full

```python
print(np.zeros((2,2)))
print(np.ones((2,2)))
print(np.full((2,2),9))
print(np.eye(3))
```

## arange & linspace

```python
print(np.arange(1,10))
print(np.arange(1,10,2))
print(np.linspace(0,1,5))
print(np.linspace(10,20,4))
```

## Random

```python
print(np.random.randint(1,10,5))
print(np.random.rand(3))
print(np.random.randn(3))
print(np.random.choice([10,20,30],5))
```

## Mathematical Functions

```python
arr=np.array([10,20,30,40])
print(np.sum(arr))
print(np.min(arr))
print(np.max(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
```

## Dot Product

```python
a=np.array([1,2,3])
b=np.array([4,5,6])
print(np.dot(a,b))
print(a@b)
```

## Matrix Multiplication

```python
a=np.array([[1,2],[3,4]])
b=np.array([[5,6],[7,8]])
print(np.matmul(a,b))
print(a@b)
```

## Sorting & Searching

```python
arr=np.array([40,10,30,20])
print(np.sort(arr))
print(np.argsort(arr))
print(np.where(arr==30))
print(np.searchsorted(np.sort(arr),25))
```

## Stacking & Splitting

```python
a=np.array([1,2])
b=np.array([3,4])
print(np.vstack((a,b)))
print(np.hstack((a,b)))
arr=np.array([1,2,3,4,5,6])
print(np.split(arr,3))
```

## Linear Algebra

```python
a=np.array([[1,2],[3,4]])
print(np.linalg.inv(a))
print(np.linalg.det(a))
vals,vecs=np.linalg.eig(a)
print(vals)
print(vecs)
```

## Pandas

```python
import pandas as pd
```

## Series

```python
s=pd.Series([10,20,30])
print(s)
print(s[0])
print(s[1:])
print(s.mean())
```

## Creating DataFrames

```python
df=pd.DataFrame({'Name':['A','B'],'Age':[20,25]})
print(df)
print(df.columns)
print(df.shape)
print(df.head())
```

## Reading Data

```python
df1 = pd.read_csv('data.csv')
df2 = pd.read_excel('data.xlsx')
df3 = pd.read_json('data.json')
```

## Writing Data

```python
df.to_csv('data.csv')
df.to_excel('data.xlsx')
df.to_json('data.json')
```

## Info & Shape

```python
df=pd.DataFrame({'Name':['A','B'],'Age':[20,25]})
df.info()
print(df.shape)
print(df.describe())
```

## loc & iloc

```python
df=pd.DataFrame({'Name':['A','B','C'],'Age':[20,25,30]})
print(df.loc[0])
print(df.loc[:,['Name']])
print(df.iloc[1])
print(df.iloc[:2])
```

## Boolean Filtering

```python
df=pd.DataFrame({'Age':[20,25,30,35]})
print(df[df['Age']>25])
print(df[df['Age']==30])
```

## Query Method

```python
df=pd.DataFrame({'Age':[20,25,30,35]})
print(df.query('Age>25'))
print(df.query('Age==30'))
```

## Missing Values

```python
df=pd.DataFrame({'A':[1,None,3]})
print(df.isnull())
print(df.notnull())
print(df.fillna(0))
print(df.dropna())
```

## Duplicates

```python
df=pd.DataFrame({'A':[1,1,2,2]})
print(df.duplicated())
print(df.drop_duplicates())
```

## Rename Columns

```python
df=pd.DataFrame({'Age':[20,25]})
print(df.rename(columns={'Age':'Student_Age'}))
```

## Change Data Types

```python
df=pd.DataFrame({'Age':['20','25']})
print(df.dtypes)
df['Age']=df['Age'].astype(int)
print(df.dtypes)
```

## Replace Values

```python
df=pd.DataFrame({'City':['Pune','Mumbai','Pune']})
print(df.replace('Pune','Nashik'))
```

## get_dummies

```python
df=pd.DataFrame({'City':['Pune','Mumbai','Delhi']})
print(pd.get_dummies(df['City']))
```

## Sorting & Ranking

```python
df=pd.DataFrame({'Marks':[50,80,70]})
print(df.sort_values('Marks'))
df['Rank']=df['Marks'].rank(ascending=False)
print(df)
```

## GroupBy

```python
df=pd.DataFrame({'City':['Pune','Pune','Mumbai'],'Sales':[100,200,300]})
print(df.groupby('City')['Sales'].sum())
print(df.groupby('City')['Sales'].mean())
```

## Pivot Tables

```python
df=pd.DataFrame({'City':['Pune','Pune','Mumbai'],'Sales':[100,200,300]})
print(pd.pivot_table(df,values='Sales',index='City',aggfunc='sum'))
```

## Date & Time

```python
dates=pd.DataFrame({'Date':['2025-01-01','2025-02-01']})
dates['Date']=pd.to_datetime(dates['Date'])
print(dates['Date'].dt.year)
print(dates['Date'].dt.month)
```

## String Operations

```python
s=pd.Series(['Alice','Bob','Charlie'])
print(s.str.upper())
print(s.str.lower())
print(s.str.contains('a',case=False))
print(s.str.len())
```
